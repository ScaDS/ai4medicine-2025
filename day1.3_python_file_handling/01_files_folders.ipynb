# Processing files and folders

Python allows to process files and folders on your operating system. In this notebook, we want to start with reading files and their properties from a folder using the built-in module [`os`](https://docs.python.org/3/library/os.html).

For more information on file handling also see:
* [RealPython - Working With Files in Python](https://realpython.com/working-with-files-in-python/)



```python
import os

path = "data_banana/"
```

A simple and straight-forward method to read the contents of a directory is [`listdir()`](https://docs.python.org/3/library/os.html#os.listdir).  

It returns a list containing the names (Strings) of the entries in the given directory.

```python
file_list = os.listdir(path)

print(type(file_list[0]))

file_list
```

**Attention**: Like most of such methods, listdir yields entries in arbitrary order!

But we can sort them alphabetically:

```python
file_list_sorted = sorted(file_list)

file_list_sorted
```

As an alternative, [`scandir()`](https://docs.python.org/3/library/os.html#os.scandir) provides objects which also contain additional information about file type or file attributes.

**Attention**: scandir yields entries in arbitrary order!

```python
with os.scandir(path) as entries:
    for entry in entries:
        if entry.is_file():
            print(f"File:\t{entry.path} - {entry.stat().st_size} bytes")
        if entry.is_dir():
            print(f"Dir:\t{entry.name}/ - {entry.stat().st_size} bytes")
```

We can also filter our file list so that we are left with image files only. This can be done via a short for statement

```python
image_file_list = [file for file in file_list_sorted if file.endswith(".tif")]

image_file_list
```

And we can show our images using additional libaries. In this case we use [`skimage`](https://scikit-image.org/) to read the actual image data from the file, and [`stackview`](https://github.com/haesleinhuepf/stackview) to display the image.

```python
from skimage.io import imread
import stackview
```

```python
# Lets display 5 of the images
for file in image_file_list[10:15]:
    image = imread(path + file)
    stackview.imshow(image)
```
